```python
import ast
import builtins
import pandas as pd
from datasets import load_dataset
```

#### Load HR MultiWOZ dataset from Huggingface

```python
ds = load_dataset("xwjzds/hr_multiwoz_tod_sgd")
```

```python
# Function to convert value as string but represented to list or dict 
def convert_string_to_list_or_dict(data_item):
    for key, value in data_item.items():
        if isinstance(value, str) and (value.startswith("[") or value.startswith("{")):
            try:
                data_item[key] = ast.literal_eval(value)
            except (ValueError, SyntaxError):
                pass  # Ignore invalid conversions
    
    return data_item
```

```python
# convert all items of dataset that have string represented to list or dict
list_ds = []
for item in ds_train:
    convert_item = convert_string_to_list_or_dict(item)
    list_ds.append(convert_item)
```

```python
# Function to convert single-element lists to strings
def convert_list_to_str(input_list):
    if isinstance(input_list, builtins.list) and len(input_list) == 1:
        return input_list[0]
    return input_list
```

```python
# Convert all single-element lists to strings in your dataset
for item in list_ds: 
    item['service'] = convert_list_to_str(item['service'])
```

```python
# Convert list of dataset to dataframe
df = pd.DataFrame(list_ds[0:], columns=list_ds[0])
```
